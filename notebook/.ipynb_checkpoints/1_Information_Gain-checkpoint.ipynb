## DECISION TREE AND INFOMATION GAIN



$Gain(A) = E(T) - E_A(T)$

Entropia di T prima di fare split

$E(T) = - \sum_{i=1}^m p_i \log_2(p_i)$

Entropia di T dopo aver diviso per il valore A


$E_A(T) = \sum_{v \in values(A)} \frac {|T_v|}{T} x E(T_v)$

#### **Training set (T)**

| age         | income     | student    | credit_rating    | compra_computer  |
|-----        |----        |----        | -------------    | -------------   |
| youth         | high     | no    | fair    | no   |
| youth         | high     | no    | excellent    | no   |
| middle_aged         | high     | no    | fair    | si   |
| senior         | medium     | no    | fair    | si   |
| senior         | low     | si    | fair    | si   |
| senior         | low     | si    | excellent    | no   |
| middle_aged         | low     | si    | excellent    | si   |
| youth         | medium     | no    | fair    | no   |
| youth         | low     | si    | fair    | si   |
| senior         | medium     | si    | fair    | si   |
| youth         | medium     | si    | excellent    | si   |
| middle_aged         | medium     | no    | excellent    | si   |
| middle_aged         | high     | si    | fair    | si   |
| senior         | medium     | no    | excellent    | no   |

- Class label: compra_computer
- Classificazione binaria -- 2 classi {si, no}
- 9 elementi di classe si and 5 tuples of class no

```python
import math
```

```python
E_T = -9/14 * math.log2(9/14) - (5/14) * math.log2(5/14)
print(f'E(T) = {E_T:.3f}')
```


- Calcolare information gain per ogni attributo, quello con il valore più alto è scelto come attributo di "split"

#### **age**

Values(age) = {youth, middle_aged, senior}
- il valore 'youth' ha 2 tuple "si" e 3 no
- 'middle_aged' ha 4 tuple "si" e 0 no
- 'senior' ha 3 tuple "si" e 2 no

```python
E_age = 5/14 * (-2/5 * math.log2(2/5) - (3/5) * math.log2(3/5)) + 4/14 * (- 4/4 * math.log2(4/4)) + 5/14 * (-3/5 * math.log2(3/5) - (2/5) * math.log2(2/5))
print(f'E_age = {E_age:.3f}')
```

```python
Gain_age = E_T - E_age
print(f'Gain(age) = {Gain_age:.3f}')
```

#### **income**
Values(income) = {high, medium, low}
- 'high': 2 si, 2 no
- 'medium': 4 si e 2 no
- 'low': 3 si e 1 no

```python
E_in = 4/14 * (-2/4 * math.log2(2/4) - 2/4 * math.log2(2/4)) + 6/14 * (-4/6 * math.log2(4/6) - 2/6 * math.log2(2/6)) + 4/14 * (-3/4 * math.log2(3/4) - 1/4 * math.log2(1/4))
print(f'E_income = {E_in:.3f}')
```

```python
Gain_in = E_T - E_in
print(f'Gain(income) = {Gain_in:.3f}')
```

#### **student**
Values(student) = {no, yes}
- per 'no' ci sono 3 tuple "si" e 4 no
- per 'yes' ci sono 6 tuple "si" e 1 no

```python
E_st = 7/14 * (-3/7 * math.log2(3/7) - 4/7 * math.log2(4/7)) + 7/14 * (-6/7 * math.log2(6/7) - 1/7 * math.log2(1/7))
print(f'E_student = {E_st:.3f}')
```

```python
Gain_st = E_T - E_st
print(f'Gain(student) = {Gain_st:.3f}')
```

#### **credit_rating**
Values(credit_rating) = {fair, excellent}
- per 'fair' 6 si e 2 no
- per 'excellent' 3 si e 3 no

```python
E_cr = 8/14 * (-6/8 * math.log2(6/8) - 2/8 * math.log2(2/8)) + 6/14 * (-3/6 * math.log2(3/6) - 3/6 * math.log2(3/6))
print(f'E_credit_rating = {E_cr:.3f}')
```

```python
Gain_cr = E_T - E_cr
print(f'Gain(credit_rating) = {Gain_cr:.3f}')
```

Age ottiene IG più alto e viene selezionato
