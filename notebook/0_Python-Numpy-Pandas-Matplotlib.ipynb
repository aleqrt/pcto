### Indexing

L'indicizzazione è il processo di accesso a un elemento in una sequenza utilizzando la sua posizione nella sequenza (il suo indice).

In Python, l'indicizzazione inizia da 0, il che significa che il primo elemento di una sequenza è nella posizione 0, il secondo elemento è nella posizione 1 e così via.

Per accedere a un elemento in una sequenza, puoi utilizzare parentesi quadre [] con l'indice dell'elemento a cui desideri accedere.

```python
my_list = ['apple', 'banana', 'cherry', 'date']
print(my_list[0]) # output: 'apple'
print(my_list[1]) # output: 'banana'
```

### Slicing


"Slicing" è il processo di accesso a una sottosequenza di una sequenza specificando un indice iniziale e finale. Python usa l'operatore ":"

```python
print(my_list[1:3]) # output: ['banana', 'cherry']
print(my_list[:2]) # output: ['apple', 'banana']
print(my_list[2:]) # output: ['cherry', 'date']
```

####Dictionaries

Un dizionario memorizza coppie (chiave, valore)
Il dict di crea con dict() o {}

```python
d = {'cat': 'cute', 'dog': 'furry'} # Crea un nuovo dizionario con alcuni dati
print(d['cat']) # Ottieni una voce da un dizionario; stampa "cute"
print('cat' in d) # Controlla se un dizionario ha una determinata chiave; stampa "true"
print('wolf' in d)
```

```python
keys = ['apples', 'oranges', 'bananas', 'cherries']
values = [3, 4, 2, 10]
fruits = dict(zip(keys, values)) # zip permette di associare gli elementi in una tupla
print(fruits)
```

### Creare ndarray

We can create ndarray from a list or tuple

```python
data1 = [1, 2, 3, 4, 5] # list
arr1 = np.array(data1) # 1-d array
data2 = [range(1, 5), range(5, 9)] # list of lists
arr2 = np.array(data2) # 2-d array
arr2.tolist() # converti array in lista
```

```python
arrFromTuple = np.array((1,'a',3.0))  # crea ndarray da una tupla
print(arrFromTuple)
```

Si possono creare array "speciali"

```python
print(np.zeros(10)) #1-d array di zero
print(np.zeros((3, 6))) #2-d array di zero
print(np.ones(10)) #1-d array di uno
```

```python
np.linspace(0, 1, 5) # 1d array from 0 to 1 (inclusive) with 5 points
```

```python
np.logspace(0, 3, 4) # 1d array from 10^0 to 10^3 (inclusive) with 4 points
```

## Matplotlib

Reference: https://matplotlib.org

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(9, 3))
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
plt.plot(x, sinus)
plt.show()
```

```python
# subplots

x = np.arange(1, 5)
y_1 = 2*x + 5
y_2 =  -2*x + 15

plt.figure(figsize=(8, 7))
# Set up a subplot grid that has height 2 and width 1, and set the first such subplot as active
plt.subplot(2, 1, 1)
plt.plot(x, y_1)
plt.title('First line')

# Set the second subplot as active, and make the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_2)
plt.title('Second line')

plt.subplots_adjust(hspace=0.3)

plt.show()
```

```python
# barplot

# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.bar(x, y)

# function to show the plot
plt.show()
```

```python
#scatter

# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot scatter
plt.scatter(x, y)

# function to show the plot
plt.show()
```

```python
#boxplot
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

# Creating plot
plt.boxplot(data)

# show plot
plt.show()
```
