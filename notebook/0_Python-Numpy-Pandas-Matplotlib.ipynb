### Control Flow Statements
- Similar to other programming languages, the control flow statements in Python include if, for, and while statements

```python
# using if-else statement

x = 15

if x % 2 == 0:
    print("x =", x, "is even")
else:
    print("x =", x, "is odd")

x = 0

# using if-elif-else statement
if x > 0:
    print("x =", x, "is positive")
elif x < 0:
    print("x =", x, "is negative")
else:
    print("x =", x, "is neither positive nor negative")
```

```python
# # using loop statement

fruits = ['apple', 'orange', 'kiwi']

for elem in fruits:
    print(elem, len(elem))
```

```python
for index in range(len(fruits)):
    print(fruits[index])
```

```python
# using while
a = list(range(-10,10))

i = 0
while (a[i] < 0):
    i = i + 1

print(i)
print("First non-negative number:", a[i])
```

### Series

- It consists of a one-dimensional array of values, whose elements can be referenced using an index array
- It can be created from a list, a numpy array, or a Python dictionary

We can apply most of the numpy functions on the Series object

```python
import pandas as pd
```

```python
s = pd.Series([2.1, 3., -1., 0.5, -5.2, 1.5])   # creating a series from a list
s
```

```python
print('Values=', s.values)     # display values of the Series
print('Index=', s.index)       # display indices of the Series
```

```python
s2 = pd.Series(np.arange(6, 20, 3))  # creating a series from a numpy ndarray
s2
```

```python
my_dict = {'name': 'Jack', 'age': 26}
s3 = pd.Series(my_dict)   # creating a series from dictionary object
s3
```

### DataFrame

- It is a tabular, containing a collection of columns, each of which can be of different types (numeric, string, boolean, etc)
- Unlike Series, a DataFrame has distinct row and column indices
- There are many ways to create a DataFrame object (e.g., from a dictionary, list of tuples, or even numpy's ndarrays).

```python
my_dict = {'names': ['Nick', 'Jess', 'Winston', 'Schmidt', 'Cece'],
        'births': [150, 200, 100, 50, 70]}
dataframe = pd.DataFrame(my_dict) # creating DataFrame from dictionary
dataframe
```

```python
# Creating DataFrame from a list of tuples
columns = ['names', 'births']
names = ['Nick', 'Jess', 'Winston', 'Schmidt', 'Cece']
births = [150, 200, 100, 50, 70]
dataset = list(zip(names, births))
df = pd.DataFrame(dataset, columns=columns)
df
```

```python
# Creating DataFrame from numpy ndarray
npdata = np.array([range(3), range(4, 7)])
columnNames = ['x1','x2','x3']
data = pd.DataFrame(npdata, columns=columnNames)
data
```

### Arithmetic Operations

```python
npdata = np.array([range(3), range(4, 7)])
columnNames = ['x1','x2','x3']
data = pd.DataFrame(npdata, columns=columnNames)
```

```python
npdata2 = np.array([range(9, 12), range(2, 5)])
data2 = pd.DataFrame(npdata2, columns=columnNames)
```

```python
print('data1 =\n', data)
print('\ndata2 =\n', data2)

print('\ndata + data2 = ')
print(data.add(data2))

print('\ndata * data2 = ')
print(data.mul(data2))
```

```python
print(data.abs())    # get the absolute value for each element

print('\nMaximum value per column:')
print(data.max())    # get maximum value for each column

print('\nMinimum value per row:')
print(data.min(axis=1))    # get minimum value for each row

print('\nSum of values per column:')
print(data.sum())    # get sum of values for each column

print('\nAverage value per row:')
print(data.mean(axis=1))    # get average value for each row
```

## Matplotlib

- It is a library for creating static, animated and interactive visualizations in Python

Reference: https://matplotlib.org

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 5)
y = 2*x + 5

plt.figure(figsize=(8, 7))
plt.plot(x, y)

plt.title('Line')

plt.xlabel('x axis')
plt.ylabel('y label')

# Show the figure
plt.show()
```

```python
x = np.arange(1, 5)
y = 2*x + 5

plt.figure(figsize=(8, 7))
plt.plot(x, y, 'or')
plt.title('Line')
plt.xlabel('x axis')
plt.ylabel('y label')
plt.show()
```

```python
x = np.arange(1, 5)
y = 2*x + 5

plt.figure(figsize=(8, 7))
plt.plot(x, y, 'x', label='First line')
plt.title('Different lines')
plt.xlabel('x axis')
plt.ylabel('y axis')

y = -2*x + 15
plt.plot(x, y, '--', label='Second line')

plt.legend()

plt.show()
```

```python
# subplots

x = np.arange(1, 5)
y_1 = 2*x + 5
y_2 =  -2*x + 15

plt.figure(figsize=(8, 7))
# Set up a subplot grid that has height 2 and width 1, and set the first such subplot as active
plt.subplot(2, 1, 1)
plt.plot(x, y_1)
plt.title('First line')

# Set the second subplot as active, and make the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_2)
plt.title('Second line')

plt.subplots_adjust(hspace=0.3)

plt.show()
```

```python
# subplots

x = np.arange(1, 5)
y_1 = 2*x + 5
y_2 =  -2*x + 15

plt.figure(figsize=(13, 10))
# Set up a subplot grid that has height 2 and width 2, and set the first such subplot as active
plt.subplot(2, 2, 1)
plt.plot(x, y_1)
plt.title('First line')

# Set the second subplot as active, and make the second plot
plt.subplot(2, 2, 2)
plt.plot(x, y_2)
plt.title('Second line')

# Set the third subplot as active, and make the third plot
plt.subplot(2, 2, 3)
plt.plot(x, y_1, 'r--')
plt.title('First line in red')

# Set the fourth subplot as active, and make the fourth plot
plt.subplot(2, 2, 4)
plt.plot(x, y_2, 'gx')
plt.title('Second line in green')

plt.subplots_adjust(hspace=.2)
plt.subplots_adjust(wspace=.2)

plt.show()
```

```python
# barplot

x = [5, 8, 10]
y = [12, 16, 6]

x2 = [6, 9, 11]
y2 = [6, 15, 7]

plt.figure(figsize=(8, 7))
plt.bar(x, y, align='center')
plt.bar(x2, y2, color='y', align='center')
plt.title('Bar graph')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
# histogram

a = np.array([1, 2, 1, 5, 6, 2, 3, 3, 8, 1, 1, 7, 8, 8, 6])
plt.figure(figsize=(8, 7))
plt.hist(a)
plt.title("histogram")
plt.show()
```
