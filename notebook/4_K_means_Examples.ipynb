##L’algoritmo K-means inizia scegliendo casualmente un valore del centroide per ogni cluster. L’algoritmo procede iterativamente:

* Trova la distanza Euclidea tra ogni istanza di dati e i centroidi di tutti i cluster;
* Assegna le istanze di dati al cluster del centroide con la distanza più vicina;
* Calcola i nuovi valori dei centroidi basati sui valori medi delle coordinate di tutte le istanze di dati dal cluster corrispondente.




**K-Means clustering on on the Iris dataset**

```
url = 'https://raw.githubusercontent.com/aleqrt/pcto-valentini/main/notebook/iris.csv'
```

**K-means**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#importing the Iris dataset with pandas
#import the dataset
df = pd.read_csv(url)
df.head(10)
x = df.iloc[:, [0,1,2,3]].values
```

```
from sklearn.cluster import KMeans

kmeans5 = KMeans(n_clusters=3)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)

kmeans5.cluster_centers_
```

```
print(kmeans5.labels_)
```

```
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans5, s=50, cmap='viridis')

centers = kmeans5.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```
Error = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    #The K-means algorithm aims to choose centroids that minimise the inertia
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
#ELbow è una tecnica utilizzata per determinare il numero ottimale di cluster in un algoritmo di clustering, come K-means
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```
